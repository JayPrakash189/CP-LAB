```python
import numpy as np
from numba import cuda

@cuda.jit
def prefix_max_kernel(inp, out):
    idx = cuda.grid(1)

    if idx < len(inp):
        max_val = inp[0]

        for i in range(idx + 1):
            if inp[i] > max_val:
                max_val = inp[i]

        out[idx] = max_val

n = 10000000

input_array = np.random.rand(n).astype(np.float32)
output_array = np.zeros_like(input_array)

d_input = cuda.to_device(input_array)
d_output = cuda.device_array_like(output_array)

threads_per_block = 256
blocks_per_grid = (n + threads_per_block - 1) // threads_per_block

prefix_max_kernel[blocks_per_grid, threads_per_block](d_input, d_output)

output_array = d_output.copy_to_host()

print("First 10 Input Values:")
print(input_array[:10])

print("\nFirst 10 Prefix Maximum Values:")
print(output_array[:10])
```
